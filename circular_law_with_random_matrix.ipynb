```python
import numpy

n = 2000
X = numpy.random.randn(n, n)
vals, vecs = numpy.linalg.eig(X)
vals = numpy.array(vals)
x = vals.real
y = vals.imag
```

```python
import matplotlib
import tabulate

graphSize = numpy.ceil(max([min(x), min(y), max(x), max(y)], key = abs)) * 1.3
matplotlib.pyplot.scatter(x, y, s=1)
matplotlib.pyplot.ylabel('imaginary')
matplotlib.pyplot.xlabel('real')
matplotlib.pyplot.axis([-graphSize, graphSize, -graphSize, graphSize])
matplotlib.pyplot.axes().set_aspect('equal') 
matplotlib.pyplot.show()

data = [{'part': 'real', 'min': min(x), 'max': max(x), 'average':numpy.average(x), 'median': numpy.median(x), 'SD': numpy.std(x)},
        {'part': 'imaginaly', 'min': min(y), 'max': max(y), 'average':numpy.average(y), 'median': numpy.median(y), 'SD': numpy.std(y)}]
print(tabulate.tabulate(data, headers='keys', numalign='left', tablefmt='github'))
```
